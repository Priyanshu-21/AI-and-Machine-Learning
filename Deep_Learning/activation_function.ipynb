### Visualization of all the activation functions 
1. `Binary Classification`: Sigmoid Function
2. Tangent Function or tanh
3. `Multi-Class Classification`: - Softmax Function
4. `Rectified Linear Unit` or ReLU
    - Leaky ReLU
    - Parametric ReLU
    - Exponential ReLU
5. `Modern Activation Functions`
    - Swish (Self - Gated Functions)
    - Mish


```python
#Importing Libraries 
import numpy as np 
import matplotlib.pyplot as plt
```

```python
#Definition of activation functions

#Define range of inputs (Randomly gettting 500 inputs in Range [-5, 5]) 
x = np.linspace(-5, 5, 500)

#Define each activation function 
def sigmoid(x): return 1.0 / (1.0 + np.exp(-x))
def tanh(x): return np.tanh(x)
def ReLU(x): return np.maximum(0, x)
def leaky_ReLU(x, alpha= 0.01): return np.where(x > 0, x, alpha * x)
def exp_ReLu(x, alpha= 1): return np.where(x > 0, x, alpha * (np.exp(x) - 1))
def swish(x): return x * sigmoid(x)
def mish(x): return x * np.tanh(np.log1p(1 + np.exp(x)))

#key-value storage of activation function for better handling 
activation_functions = {
    'Sigmoid': sigmoid,
    'Tangent': tanh,
    'Rectified Linear Unit': ReLU,
    'Leaky ReLU': leaky_ReLU,
    'Exponential ReLU': exp_ReLu,
    'Swish Func': swish,
    'Mish Func': mish,
}
```

```python
#Visualization of Each activaiton function 
plt.figure(figsize= (12, 16))

for i, (name, func) in enumerate(activation_functions.items(), 1):
    
    #Sub-Plot
    plt.subplot(4, 2, i)
    y = func(x)

    plt.plot(x, y, label= name, linewidth= 2)
    plt.axhline(color= 'black', linewidth= 0.8)
    plt.axvline(color= 'black', linewidth= 0.8)
    plt.title(label= name)
    plt.grid(visible= True, linestyle= '--', alpha= 0.6)

plt.show()
```
