### Introduction to K-Means clustering

K-means is vastly used for clustering in many data science applications. It is especially useful if you need to quickly discover insights from unlabeled data.

- Real-world applications of k-means include:   
    - Customer segmentation
    - Understanding what website visitors are trying to accomplish
    - Pattern recognition
    - Feature engineering
    - Data compression

### Understanding use of K-Means clustering problem in a randomly generated clusters
- Randomly generate the datasets 
- Choose clusters and partition the datasets 
- Use K-Mean method to cluster the dataset and model development 

```python
#Importing libraries 
import numpy as np
import pandas as pd 
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt 
```

```python
#np.random.seed(0) makes NumPyâ€™s random results predictable and reproducible.
np.random.seed(0)
```

```python
#Using make_blobs method to make 4 clusters 
x, y = make_blobs(n_samples= 5000, centers= [[4, 4], [-2, -1], [2, -3], [1, 1]], cluster_std= 0.9)
```

```python
#Visualization of randomly generated values 
plt.scatter(x[:, 0], x[:, 1], ec= 'k', alpha= 0.6, s= 60, marker= '.')
plt.show()
```

### Applying k-Means clustering algorithm in this model 
The KMeans class has many parameters that can be used, but we will be using these three:

- `init`: Initialization method of the centroids.
  - Value will be: `k-means++` 
  - `k-means++`: Selects initial cluster centres for k-means clustering in a smart way to speed up convergence.
- `n_clusters`: The number of clusters to form as well as the number of centroids to generate. 
  -  Value will be: 4 (since we have 4 centres) 
- `n_init`: Number of times the k-means algorithm will be run with different centroid seeds. The final results will be the best output of n_init consecutive runs in terms of inertia. 
   - Value will be: 12 

```python
#Model Development 
from sklearn.cluster import KMeans

kMean = KMeans(init= 'k-means++', n_clusters= 4, n_init= 12)
```

```python
kMean.fit(x)
```

```python
#Get labels and cluster_centroids 
k_means_labels = kMean.labels_
k_means_labels

k_means_clusters = kMean.cluster_centers_
k_means_clusters
```

```python
#Visualization of K-Means algorithm prediction and see how well it clustered the datasets 
#Plot dimentions 
fig = plt.figure(figsize= (10, 8))

#Assigns distinct colors to each cluster label in k_means_labels.
#Different color values to clusters (Each color corresponds to a cluster -> RGBA)
colors = plt.cm.tab10(np.linspace(0, 1, len(set(k_means_labels))))

ax = fig.add_subplot(1, 1, 1)
#Make Scatterplot with each cluster and centroid value's with defined colors 
for k, col in zip(range(len([[4, 4], [-2, -1], [2, -3], [1, 1]])), colors):
    
    #creates a boolean mask that selects points belonging to cluster k
    my_members = (k_means_labels == k)
    
    #make cluster centroids 
    cluster_centroids = k_means_clusters[k]

    #Plotting datapoints scatterplot 
    ax.plot(x[my_members, 0], x[my_members, 1], 'w', markerfacecolor= col, marker= '.', ms= 10)

    #Plotting cluster centroids 
    ax.plot(cluster_centroids[0], cluster_centroids[1], 'o', markerfacecolor= col, markeredgecolor= 'k', markersize= 6)

plt.title('K - Means Clustering')
plt.tight_layout()
plt.show()
```

### Analysis of K-Mean's Clustering with different k (clusters values)  
1. Trying with 3 clusters 
2. Trying with 5 clusters 

```python
#Model Development with k = 3
k_mean_3 = KMeans(init= 'k-means++', n_clusters= 3, n_init= 12)
k_mean_3.fit(x)
```

```python
#Get labels and clusters 
k_mean_3_labels = k_mean_3.labels_
k_mean_3_clusters = k_mean_3.cluster_centers_
```

```python
#Model Visualization 
fig = plt.figure(figsize= (12, 8))

#Set of colors for different clusters 
colors = plt.cm.tab10(np.linspace(0, 1, len(set(k_mean_3_labels))))

ax = fig.add_subplot(1, 1, 1)

#Plotting datapoint and cluster of K-Means Model 
for k, col in zip(range(len([[4, 4], [-2, -1], [2, -3]])), colors):

    #Masking cluster with datapoints 
    my_members_3 = (k_mean_3_labels == k)

    #Slecting cluster centroids 
    cluster_centroids_3 = k_mean_3_clusters[k]

    #Plot datapoint in the graph
    ax.plot(x[my_members_3, 0], x[my_members_3, 1], 'w', markerfacecolor= col, marker= '.', ms= 10)

    #Plot cluster centroids value 
    ax.plot(cluster_centroids_3[0], cluster_centroids_3[1], 'o', markerfacecolor= col, markeredgecolor= 'k', markersize= 8)

plt.title("K Means with 3 Clusters")
plt.tight_layout()
plt.show()
```

```python
#Model Development with k = 5
k_mean_5 = KMeans(init= 'k-means++', n_clusters= 5, n_init= 12)
k_mean_5.fit(x)
```

```python
#Get labels and clusters 
k_mean_5_labels = k_mean_5.labels_
k_mean_5_clusters = k_mean_5.cluster_centers_
```

```python
#Visualization of 5 clusters
fig = plt.figure(figsize= (10, 8))

#Color Map to map out differnt clusters 
colors = plt.cm.tab10(np.linspace(0, 1, len(set(k_mean_5_labels))))

ax = fig.add_subplot(1, 1, 1)
#Visualization 
for k, col in zip(range(len([[4, 4], [-2, -1], [2, -3], [1, 1], [5, 5]])), colors):

    #bit Masking to get each cluster datapoints 
    my_members_5 = (k_mean_5_labels == k)

    #Get the centroid of each cluster
    cluster_centroid_5 = k_mean_5_clusters[k]

    #Plot of datapoints of 5 clusters 
    ax.plot(x[my_members_5, 0], x[my_members_5, 1], 'w', markerfacecolor= col, marker= '.', ms=10)

    #Plot of centroid's of each cluster 
    ax.plot(cluster_centroid_5[0], cluster_centroid_5[1], 'o', markerfacecolor= col, markeredgecolor= 'k', markersize= 8)

plt.title("K Means with 5 clusters")
plt.tight_layout()
plt.show()
```
